# Cours 6 : Sanity check : Physics vs Reality

The goal of this course is to test the **methods** that we learned and investigate if they can help us to capture real world motions.

The motion we try to elaborate on is the **projectile motion** applied to the field of **sport engineering**. There is a plethora of sports based on the projectile motion: football, basket ball, base ball, golf etc. In all these sports, tracking precisely the ball's position can be really crucial. For example, let us focus on **tennis** where no less than 10 high definition cameras track the ball all around the court. The ball's position is captured at 340 frame per second. The time interval between two measurements (two frames) - $\Delta t$ (see course 2) - is given by $1/340\approx 3$ milliseconds. Such precise measurements allows to capture the initial conditions at the instant when the player hits the ball. From these conditions, the ball's trajectory can be determined by **modeling** the forces acting on the ball and using **Newton's second law**.

The question is the following: what is the **maximal velocity** $v_{max}$ that the ball can have and still remain in the court at the end of its trajectory?  

### Temptative answer #1 : gravity

To get a rough estimate, one could consider only the effect of gravity and neglect all other forces. A velocity is a distance $d$ divided by a time $t$. The distance $d$ to consider is the maximal range one could reach to remain in the play: $d=11.885$ m. To compute the time, one knows that the time required to touch the floor in the projectile motion is the same than for the motion with no horizontal component (recall https://www.youtube.com/watch?v=zMF4CD7i3hg). 

The net has a height $y_{0}$ of 1.065 m, defining the time $t$ through:

$$
y_{0}=\frac{gt^{2}}{2}
$$

Eventually, we have:

$$v_{max}=\frac{d}{t}=\frac{d}{\sqrt{\frac{2y_{0}}{g}}}$$

Which one can estimate (with a calculator) to be around 25.5 [m/s] or 91.8 [km/h]. 

Is this the actual physical limit ? Let us have a look to the world record for a forehand. It is Andy Murray who holds the world record for the fastest forehand ever hit at an (approximately horizontal) speed of 199.5587 km/h. With no other force than gravity (oriented along $\textbf{e}_{y}$), the horizontal component of the velocity has no time dependence and the ball crosses the net with the same horizontal velocity. 

What **distance** would the ball cover, in the aforementioned problem, for an initial velocity of 55.43 m/s (199.5 km/h) (https://www.youtube.com/watch?v=f1vkkqoTTPc).

We have:

$$d=55.43*\sqrt{\frac{2y_{0}}{g}}\approx 25.84$$

You have to know that the court is 23.77 meters. Let us have a look to this record (this is the magic of youtube):

https://www.youtube.com/watch?v=f1vkkqoTTPc

What strikes you ? He hits the ball bellow the horizontal ($\theta_{0}<0$) **and** he is inside the court !!!

### Temptative answer #2 : gravity + drag

Now that we found that only gravity could **not** explain the way we play tennis, one will search to find the extra ingredients needed to get satisfying results. 

The first extra ingredient is the **drag force** noted $F_d$. It is illustrated on the following figure:

<img title="a title" alt="Alt text" src="../Figures/Cours 6/drag.jpg" width="500">

<div class="alert alert-block alert-danger">
The arrow depicted on this figure have different physical origin (and dimension) so never, <b>ever</b>, should you try to sum them</div>

When an object with velocity $v=|\vec{v}|$ is moving is a fluid (such as **air** for our example), it **has to displace** the fluid it travels in and, by **Newton's third principle**, if the ball exerts a force on the fluid, the fluid exerts a force on the ball that has an equal magnitude and acts in the opposite direction. 

To capture the expression of this force, let us use **dimensional analysis**. Think first on what this force could depend on.

This force depends on:
- The velocity of the ball $V$ that has the physical dimension [$LT^{-1}$] ([$m/s$])
- The fluid of the fluid $\rho$  [$ML^{-3}$] ([$kg/m^{3}$]).
- The area $A$ facing the fluid [$L^{2}$]

<div class="alert alert-block alert-info">
If you think that this force should depend on the mass of the ball itself, think of the following situation. Suppose now that the ball is "magically" fixed in the air and that it cannot move. We now blow air on the ball to realize the same air flow profile than the one depicted on the figure: the force acting on the ball should be exactly the same: $F_d$. The air does only interact with the <b>surface</b> of the ball and nothing inside the surface is relevant: Even if the ball was empty, one would have exactly the same air flow, <b>hence</b>, the <b>same</b> force</div>

One need to **construct** an object that has the physical dimension of a force [$MLT^{-2}$] with the **density**, the **velocity** and the impinging **area**

$$[F_{d}]=[\rho]^{a}[v]^{b}[A]^{c}\leftrightarrow[MLT^{-2}]=[ML^{-3}]^{a}[LT^{-1}]^{b}[L^{2}]^{c}$$

Solving for $(a,b,c)$, thanks to:

\begin{cases}
a=1 \\
-3a+b+2c = 1 \\
-b=-2 \\
\end{cases}

We find $(a,b,c)=(1,2,1)$.

The **drag force** is thus expressed as:

$$F_{d}\propto\rho v^{2}A$$

To get the exact formula, one needs to consider the **adimensional quantity** called the **drag coefficient** ($c_d$). This coefficient takes into account the nature of the surface (e.g. fluffy for the tennis ball) and the shape associated to the surface ($A=\pi R^{2}$ for the tennis ball).

<div class="alert alert-block alert-danger">
For a sphere, the area impinging is just the projected area of the sphere on a plane: $\pi R^{2}$ and <b>not</b> the total sphere area $4\pi R^{2}$</div>

The complete formula is the following (the factor $\frac{1}{2}$ stands there for historical reasons):

$$F_{d}=\frac{1}{2}c_{d}\rho v^{2}A$$

On the following figure, you can find the **measured** drag coefficient for different shapes. This quantity is of tremendous importance in engineering (aerospatial, car industry, etc.).

<img title="a title" alt="Alt text" src="../Figures/Cours 6/drag_coef.jpg" width="300">

Also, to illustrate the importance of this drag coefficient, here is a valuable figure retrieved from Chrysler Motor Company (automotive sector).

<img title="a title" alt="Alt text" src="../Figures/Cours 2/drag_car.jpg" width="700">

Analytical computations can rapidly get complicated when this force is at play because of the **square** dependency on the velocity (leading to a non-linear differential equation). 

In the scope of this course, let us detail a direct consequence of this force on falling object: eventually, the velocity of objects subjected to both gravity and drag, as they fall from top to bottom in a straight line, reaches a **steady** velocity called the **terminal velocity** ($v_{T}$).

Terminal velocity is the maximum velocity attainable by an object as it falls through a fluid (air is the most common example). It occurs when the sum of the drag force ($F_d$) is equal to the downward force of gravity (mg) acting on the object. Since the net force on the object is zero, the object has zero acceleration. The terminal velocity is defined by the following equilibrium equation:

$$mg=\frac{1}{2}c_{d}\rho v_{T}^{2}A$$

Giving us:

$$v_{T}=\sqrt{\frac{2mg}{\rho Ac_{d}}}$$

At this point, the sum of the force being zero, the motion is characterized by a constant velocity. Here is some interesting consequence: https://www.youtube.com/watch?v=0m5CtIknAg4.

Now that we are more familiar with drag, let us get back to our simulation and change the acceleration according to the new forces at play (represented on this figure).

<img title="a title" alt="Alt text" src="../Figures/Cours 6/drag_trigo.jpg" width="600">

We have, for the net force:

$$\vec{F}=\sum_{i}\vec{f}_{i}=-F_{d}cos(\theta)\textbf{e}_{x}+(F_{d}sin(\theta)-mg)\textbf{e}_{y}$$

with

$$F_{d}=\frac{1}{2}c_{d}\rho v^{2}A$$

What is the area to consider with a tennis ball ? To find the answer, let us find out the tennis ball characteristics in this article : https://en.wikipedia.org/wiki/Tennis_ball. Let us consider the radius to be 3.3 cm.

What is the drag coefficient of a tennis ball ? This question is much more difficult to answer due to the "fluffy" surface of a tennis ball. A good estimate could be to say that it is a sphere, so, according to the above figure, on should be around 0.47. Yet, it is not the good value due to the roughness of the ball. In the following scientific paper (Cross, R., & Lindsey, C. (2014). Measurements of drag and lift on tennis balls in flight. Sports Engineering, 17(2), 89-96.), one can find $c_{d}=0.6$.

In our python code, we need to update $(a_{x},a_{y})=(\frac{dv_{x}}{dt},\frac{dv_{y}}{dt})=(0,-9.8)$ to:

$$(a_{x},a_{y})=(\frac{F_{d}}{m}cos(\theta),\frac{F_{d}}{m}sin(\theta)-g)$$

where $m$ is the mass of a tennis ball (considered as 58 g). 

To ease the notations, let us define **for the tennis ball** the constant $k$ as:

$$k_{d}=\frac{1}{2m}\rho A\approx 0.021$$

Where the **air density** $\rho$ is taken as 1.2 $[kg/m^{3}]$

We have:

$$(a_{x},a_{y})=(k*c_{d}v^{2}cos(\theta),(k*c_{d}v^{2}sin(\theta)-g)$$

Remark that:

$$v^{2}cos(\theta)=v\;v_{x}$$

and 

$$v^{2}sin(\theta)=v\;v_{y}$$

We eventually have:

$$(a_{x},a_{y})=(k*c_{d}v\;v_{x},(k*c_{d}v\;v_{y}-g))$$

Which is really compact and elegant.

The exact expression of this force is the following:

$$F_{l}=\frac{1}{2}c_{l}\rho v^{2}A$$

Exactly like $F_d$ except that the drag coefficient is replaced by the **lift coefficient** $c_l$.
