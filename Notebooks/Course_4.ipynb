To go further in the analysis, one can define the following objects: an **instanteneous velocity** as well as an **instanteneous accelaration** - derived, as in course 2, from the **average velocity** and **average acceleration** (think this through as an exercise !). 

In order to define these objects, let us define the natural directions associated with these coordinates with infinitesimal $dr$ and $d\theta$ (as it was done in course 3 for the cartesian coordinate system).

The unit vector $\textbf{e}_{r}$, at point $P$, points along the direction where **only $r$** varies. The unit vector $\textbf{e}_{\theta}$ at point $P$ points along the direction where **only $\theta$** varies. The "problem", as it can be seen on the following figure, is that these unit vector are **position dependent** (see the unit vectors defined at the black, green, yellow and purple points). Yet, this "position dependence" **does not** prevent this coordinate system from being interesting. 

<img title="a title" alt="Alt text" src="../Figures/Cours 3/polar_coord_full.jpg" width="400">

So let us go further, we have:

$$\textbf{e}_{r}=\frac{\partial\vec{P}}{\partial r}$$

and

$$\textbf{e}_{\theta}=\frac{\partial\vec{P}}{\partial \theta}$$

Let us express them, thanks to the **chain rule** in the cartesian coordinate system:

$$\textbf{e}_{r}=\frac{\partial\vec{P}}{\partial r}=\frac{\partial\vec{P}}{\partial x}\frac{dx}{dr}+\frac{\partial\vec{P}}{\partial y}\frac{dy}{d\theta}=\textbf{e}_{x}\frac{dx}{dr}+\textbf{e}_{y}\frac{dy}{dr}$$

and

$$\textbf{e}_{\theta}=\frac{\partial\vec{P}}{\partial \theta}=\frac{\partial\vec{P}}{\partial x}\frac{dx}{d\theta}+\frac{\partial\vec{P}}{\partial y}\frac{dy}{d\theta}=\textbf{e}_{x}\frac{dx}{d\theta}+\textbf{e}_{y}\frac{dy}{d\theta}$$

**Infinitesimally**, we have to acknowledge the varying $\theta$ around the circle is **equivalent**, at the limit, to vary $\theta$ along the tangent line. So we can consider the following figure to express $dx/d\theta$ and $dy/d\theta$:

<img title="a title" alt="Alt text" src="../Figures/Cours 3/infinitesimal_changes.jpg" width="400">

We have eventually:

$$\textbf{e}_{r}=cos(\theta)\;\textbf{e}_{x}+sin(\theta)\;\textbf{e}_{y}$$

$$\textbf{e}_{\theta}=-sin(\theta)\;\textbf{e}_{x}+cos(\theta)\;\textbf{e}_{y}$$

<div class="alert alert-block alert-danger">
Attention ! This time, $\theta$ is a coordinate ! Not a fixed angle relating only two coordinate systems as it was the case with $\textbf{e}_{x'}$ and $\textbf{e}_{y'}$ (see course 3)</div>

It can be easily showed that $\textbf{e}_{r}\cdot\textbf{e}_{r}=1$, $\textbf{e}_{\theta}\cdot\textbf{e}_{\theta}=1$ and $\textbf{e}_{r}\cdot\textbf{e}_{\theta}=\textbf{e}_{\theta}\cdot\textbf{e}_{r}=0$.

We can now express the position **vector** $\vec{r}$ thanks to the coordinates $(r,\theta)$ and the unit vectors $\textbf{e}_{r}$, $\textbf{e}_{\theta}$. 

By definition:

$$\vec{r}=r\;\textbf{e}_{r}$$

<div class="alert alert-block alert-danger">
The position vector cannot be express (as many student wrongly think) $\vec{r}=r\;\textbf{e}_{r}\color{red}{+\theta\;\textbf{e}_{\theta}}$. The $\theta$ dependency is already taken into account through $\textbf{e}_{r}$</div>

This formula can explicitate the link between polar and carthesian coordinates. The $x$ and $y$ coordinates of the position vector - given by:

$$x=\vec{r}\cdot\textbf{e}_{x}=rcos(\theta)$$

$$y=\vec{r}\cdot\textbf{e}_{y}=rsin(\theta)$$

from which we have the relations between the coordinates (2 equations in 2 dimensions):

$$x^{2}+y^{2}=r^{2}(cos^{2}(\theta)+sin^{2}(\theta))=r^{2}$$

and 

$$\frac{y}{x}=\frac{sin(\theta)}{cos(\theta)}=tan(\theta)$$

# Kinematics in 2D using polar coordinates

Back to the ant positions observations (above table), we can see that as the ant moves, the $r$ and $\theta$ coordinates vary with time. So we can acknowledge that, in kinematics, $r$ and $\theta$ depend on time and are actually **functions** of time: $r(t)$ and $\theta(t)$. Note that $\textbf{e}_{r}$ and $\textbf{e}_{\theta}$ also change with time (indeed, it changes from point to point). 

The position vector is then:

$$\vec{r}(t)=r(t)\;\textbf{e}_{r}(\theta(t))$$

If you can **interpret** confidently this equation, you are **fine** for the rest of the course (and bravo !). Otherwise, try your best and then, don't hesitate to come to me for questions. 

The rest is simple mathematics. 

### Instantaneous velocity

Thanks to the chain rule, we have: 

$$\vec{v}=\frac{d\vec{r}}{dt}=\frac{dr(t)}{dt}\;\textbf{e}_{r}(\theta(t))+r(t)\frac{d\textbf{e}_{r}(\theta(t))}{dt}$$

How can we differentiate $\textbf{e}_{r}(\theta(t))$ ? By expressing it in a coordinate system that **does not** depend on the position:

$$\frac{d\textbf{e}_{r}(\theta(t))}{dt}=\frac{d}{dt}(cos(\theta(t))\;\textbf{e}_{x}+sin(\theta(t))\;\textbf{e}_{y})=-\frac{d\theta(t)}{dt}sin(\theta(t))\;\textbf{e}_{x}+\frac{d\theta(t)}{dt}cos(\theta(t))\;\textbf{e}_{y}=\frac{d\theta(t)}{dt}(-sin(\theta)\;\textbf{e}_{x}+cos(\theta)\;\textbf{e}_{y})=\frac{d\theta(t)}{dt}\;\textbf{e}_{\theta}$$

Eventually, we have:

$$\vec{v}=\frac{dr(t)}{dt}\;\textbf{e}_{r}+r(t)\frac{d\theta(t)}{dt}\;\textbf{e}_{\theta}$$

The radial component $(v_{r}=\vec{v}\cdot\textbf{e}_{r})$ of the velocity is called... the **radial velocity** and the angular component ($(v_{\theta}=\vec{v}\cdot\textbf{e}_{\theta})$) is called the **angular velocity**. The angular velocity is so ubiiquitous in physics that it "has" its own greek letter $\omega=\frac{d\theta(t)}{dt}$.

To lighten a bit the notation, in mechanics, one often use an upper "dot" symbol above a quantity to signify the time derivative. We have:

$$\vec{v}(t)=\dot{r}\;\textbf{e}_{r}+r\omega\;\textbf{e}_{\theta}$$

### Instantaneous acceleration

From the instantaneous velocity, one can get the instantaneous acceleration:

$$\vec{a}=\frac{d\vec{v}}{dt}$$

The actual computation is left as an exercise but the final result is:

$$\vec{a}(t)=\frac{d\vec{v(t)}}{dt}=(\frac{d^{2}r(t)}{dt}-r(t)(\frac{d\theta(t)}{dt})^{2})\;\textbf{e}_{r}+(2\frac{dr(t)}{dt}\frac{d\theta(t)}{dt}+r(t)\frac{d^{2}\omega(t)}{dt})\;\textbf{e}_{\theta}$$

or,

$$\vec{a}(t)=\dot{\vec{v}}=(\ddot{r}-r\omega^{2})\;\textbf{e}_{r}+(2\dot{r}\dot{\omega}+r\ddot{\omega})\;\textbf{e}_{\theta}$$

# Uniform Circular Motion

The uniform circular motion is ubiquitous in physics and engineering. With polar coordinates, as one can guess, the equations of motion one have to deal with, are way easier.

The word "uniform" refers to a motion with a **constant radius** ($R$) and a constant **angular velocity** ($\omega$). A constant angular velocity leads to the following equation:

$$\omega(t)=\omega^{*}=\frac{d\theta}{dt}$$

where $\omega^{*}$ is a constant defined by $\omega^{*}=\frac{2\pi}{T}$ with $T$, the time period to complete a **full turn**. For example, the **minute hand** makes a full turn ($2*\pi$) in 60 seconds. So, 

$$\omega^{*}=\frac{2\pi}{60}\approx 0.1\;[s^{-1}]$$

The angular velocity can be **represented** with a vector that points along the axis of rotation so for the minute hand, that is turning... clockwise, it is point along $-\textbf{e}_{z}$ (with the aforementioned magnitude).

By integrating this equation, one finds an *affine* dependence for $\theta(t)$:

$$\theta(t)=\theta_{0}+\omega^{*}t$$

Often, $\theta_{0}$ (the initial angular value) is taken to be zero and the relation becomes linear (to the mathematical sense).

<div class="alert alert-block alert-info">
    Note that for a uniform <b>linear</b> motion, we had the completely analogous equation:

$$x(t)=x_{0}+vt$$
</div>

Substituting $R$ and $\omega^{*}$ into the equation obtained for the instantaneous velocity and acceleration (in polar coordinates), one finds (do the exercise):

$$\vec{r}(t)=R\;\textbf{e}_{r}$$

$$\vec{v}(t)=R\omega^{*}\;\textbf{e}_{\theta}$$

and

$$\vec{a}(t)=-R(\omega^{*})^{2}\;\textbf{e}_{r}$$

# Dynamics: Newton's law of motion

After studying tools to describe the motion of objects (kinematics), we dig now into deeper questions, namely, the relationship between the motion and the objects causing them: **forces** and **torques**. 

The first crutial observation is that those objects can be **represented** by vectors. The consequence is that the properties of vectors are valid when dealing with forces or torques.

Let us start by studying **forces**.

For example, consider the following force $\vec{f}$ (considered in 2D) represented by the vector whose components are given by the couple $(f_{x},f_{y})$ in the carthesian coordinate system defined by the unit vectors $\textbf{e}_{x}$,$\textbf{e}_{y}$. We can write:

$$\vec{f}=f_{x}\;\textbf{e}_{x}+f_{y}\;\textbf{e}_{y}$$

A force, contrary to vectors, comes with a so-called **point of application**.

When two forces act, at the same time, at a point $A$ of a body, one can show that it has the same effect as a single force (blue vector on the figure) equal to the **vector sum**:

$$\vec{R}=\vec{f_{1}}+\vec{f_{2}}$$

that is to say,

$$(R_{x},R_{y})=(f_{1x}+f_{2x},f_{1y}+f_{2y})$$

(with $f_{1x}=\vec{f_{1}}\cdot\textbf{e}_{x}$, ...)

<img title="a title" alt="Alt text" src="../Figures/Cours 4/resultante.jpg" width="300">

The vector sum of all the forces acting on the same time in a single acting point is called the **net force**. In general, for an arbitrary number of forces, the summation symbol is used:

$$\vec{R}=\sum_{i}\vec{f_{i}}$$

In addition, we know that these components transform, as the coordinates themselves, under a rotation of an angle $\theta$ of the coordinate system:

<div class="alert alert-block alert-info">
$$R_{x'}=R_{x}\;cos(\theta)+R_{y}\;sin(\theta)$$

and similarly,

$$R_{y'}=-R_{x}\;sin(\theta)+R_{y}\;cos(\theta)$$ </div>

The unit in which a force is expressed in the SI units is the **Newton** [$N=kg\cdot m\cdot s^{-2}$], that has the physical dimension $[MLT^{-2}]$. 

To **properly** analyse the motion, one need to establish **ground rules** called the **Newton's laws of motion**.

These are three basic laws of classical mechanics that describe the relationship between the motion of an object and the forces acting on it. These laws can be paraphrased as follows: 

#### First law:

**A body remains at rest, or in motion at a constant speed in a straight line, unless acted upon by a force.**

Let us comment this principle with the following picture. Suppose that the person is standing in a train on **frictionless** rollers and you record observations (as in course 2) of the position of the center of mass of this person. These observations are realised with respect to the coordinate system whose origin (**O**) is located in the corner of the carriage. Physically, we **know** that there is no **net force** acting on the person. Nevertheless, your observations shows you that, as the train accelerate, the velocity of the person is changing. A proper analysis require a frame of reference located on the rails (that can be idealized as an extremelly good approximation of **inertial** for many practical purposes). With this reference, you see that the person is indeed **at rest**.

<img title="a title" alt="Alt text" src="../Figures/Cours 4/inertial_frame_1.jpg" width="500">

As anticipated, this law helps you to define what is called an **inertial frame of reference**: a frame of reference whose origin is not undergoing any acceleration.

Here is another example of **wrong** choice:

<img title="a title" alt="Alt text" src="../Figures/Cours 4/inertial_frame_3.jpg" width="300">

Here, the coordinates of the person in the reference frame would show a complex "law of the motion". In an inertial frame althought, we can see that the person is also **at rest**.

#### Third law:

**If two bodies exert forces on each other, these forces have the same magnitude but opposite directions.**

#### Second law:

**When a body is acted upon by a force, the time rate of change of its linear momentum equals the net force.**

In classical mechanics, the **linear momentum** is defined as the following **vector**:

$$\vec{p}=m\vec{v}$$

where $m$ is the mass of the studied body and $\vec{v}$, its **instantaneous** velocity. 

The law expresses:

$$\frac{d\vec{p}}{dt}=\sum_{i}\vec{f_{i}}$$

This law is consistent with the first law since, when no net force is at play, 

$$\frac{d\vec{p}}{dt}=\frac{d(m\vec{v})}{dt}=0$$

The mass being frequently not changing with time (except e.g. for "rocket" science), so it justifies to pull $m$ out of the time derivative for most cases. Also, the "0" in this expression is a "vectorial" zero.

$$(m\frac{dv_{x}}{dt},m\frac{dv_{y}}{dt},m\frac{dv_{z}}{dt})=(0,0,0)$$

<div class="alert alert-block alert-info">
From now, we will make an extensive use of the <b>time integration</b> as it is the reciprocal operation of <b>time derivative</b> and can serve us to express velocity and position thanks to a prescribed acceleration.

$$\vec{v}(t_{2})-\vec{v}(t_{1})=\int_{\vec{v}(t_{1})}^{\vec{v}(t_{2})}d\vec{v}=\int_{t_{1}}^{t_{2}}\vec{a}dt$$

together with 

$$\vec{x}(t_{2})-\vec{x}(t_{1})=\int_{\vec{x}(t_{1})}^{\vec{x}(t_{2})}d\vec{x}=\int_{t_{1}}^{t_{2}}\vec{v}dt$$ 

Warning: A frequent source of mistake is to forget the <b>integration constants</b>

</div>

By integrating between a reference time $t_{0}$ and time $t$, this formula gives us:

$$\int_{t_{0}}^{t}\frac{d\vec{v}}{dt}dt=\int_{\vec{v}(t_{0})}^{\vec{v}(t)}d\vec{v}=\vec{v}(t)-\vec{v}(t_{0})=0$$

That is to say:

$$\vec{v}(t)=\vec{v}(t_{0})$$

All the components of the velocity are a constants of the motion. When no force is acting on a body, no component of the velocity changes with time. 

The fact that $m\frac{d\vec{v}}{dt}=0$ is also telling us some valuable additionnal pieces of informations. 

Let us use the scalar product with the velocity $\vec{0}$ on both side of the expression. We have:

$m\vec{v}\cdot\frac{d\vec{v}}{dt}=0$

Which gives:

$$mv_{x}\frac{dv_{x}}{dt}+mv_{y}\frac{dv_{y}}{dt}+mv_{z}\frac{dv_{z}}{dt}=0$$

But, we can cleverly remark that, e.g.:

$$v_{x}\frac{dv_{x}}{dt}=\frac{1}{2}\frac{d}{dt}(v_{x}^{2})$$

So the expression becomes:

$$\frac{m}{2}\frac{d}{dt}(v_{x}^{2}+v_{y}^{2}+v_{z}^{2})=\frac{d}{dt}(\frac{m|\vec{v}|^2}{2})=0$$

The **scalar** quantity $\frac{m|\vec{v}|^2}{2}$ **does not evolve with time** (in the specific situation where **no forces** are applied). Newton's second law already tells us that this quantity could be worth analysing. 

Eventually, we can now write the well-know expression: **(net) force equals mass times acceleration**. You see that to get this expression, approximations are already at play. 

$$\frac{d\vec{p}}{dt}=m\frac{d}{dt}\vec{v}=m\frac{d}{dt}\frac{d}{dt}\vec{r}=m\frac{d^{2}\vec{r}}{dt^{2}}=m\vec{a}=\sum_{i}\vec{f_{i}}$$

where $\vec{r}$ is the position vector, and $\vec{a}$ is the **instantaneous** acceleration. 

Sometimes, it is also simply rewritten:

$$\vec{a}=\frac{1}{m}\sum_{i}\vec{f_{i}}$$

# Torques and angular momentum

It is not because $\sum_{i}\vec{f_{i}}=0$ that nothing moves. Let us consider the following example:  A rod that can rotate in space around a fixed middle point (2D motion).

<img title="a title" alt="Alt text" src="../Figures/Cours 4/pure_moment.jpg" width="600">

The forces $\vec{f_{1}}$ and $\vec{f_{2}}$ are such that $\vec{f_{1}}=-\vec{f_{2}}$. From this, it is clear that $\sum_{i}\vec{f_{i}}=0$. From this figure, it seems clear that, for rotational motin, the coordinate of interest are the polar coordinates 

For example, one could study tools to understand the law of the motion for the $\theta(t)$ coordinate.

The first time derivative of $\theta$ is the **instantaneous angular velocity**. And the second derivative is the **instantaneous angular acceleration**. 

For rotational motion, what is relevant to write down the law of the motion, is the notion of **torque**.  A torque is the rotational equivalent of linear force. It is also referred to as the **moment**, **moment of force**. It represents the capability of a force to produce change in the rotational motion of the body. The concept originated with the studies by Archimedes of the usage of levers, which is reflected in his famous quote: "Give me a **lever** and a place to stand and I will move the Earth". Just as a linear force is a push or a pull, a torque can be thought of as a twist to an object around a **specific axis**.

In the same way forces can be represented with vectors, **torques** are defined thanks to the so-called **cross product** (a.k.a vector product). The scalar product "produces" a scalar and the vector product "produces" a vector. 

<div class="alert alert-block alert-info">
The vector product (or cross product), producing a vector colinear with the vector  ($\vec{n}$) normal to the plane defined by $\vec{a}, \vec{b}$: $$\vec{a}\times\vec{b}=(a_{x}b_{y}-a_{y}b_{x})\vec{n}$$  
There exists a geometrical interpretation of the cross product whose norm is the area of the parallelogram defined by $\vec{a}$ and $\vec{b}$ (exercise).
</div>

The torque is a (pseudo)vector (no need to get the significance of "pseudo" here); for point particles, it is given by the cross product of the position vector ($\vec{r}$) and the force vector. The magnitude of torque of a rigid body depends on three quantities: the force applied, the lever arm vector connecting the point about which the torque is being measured to the point of force application, and the angle between the force and lever arm vectors. In symbols:

$$\vec{\tau}=\vec{r} \times \vec{f}=|\vec{r}||\vec{f}|sin(\phi)\textbf{e}_{z}$$

where

$\vec{\tau}$ is the torque vector, $\vec{r}$ is the position vector (a vector from the point about which the torque is being measured to the point where the force is applied), $\vec{f}$ is the force vector, $\times$  denotes the cross product, which produces a vector that is perpendicular to both $\vec{r}$ and $\vec{f}$ following the **right-hand rule** (do never use your left hand - frequent mistake), $\phi$ is the angle between the force vector and the lever arm vector. The SI unit for torque is the newton-metre $[N⋅m]$. 

There exists a simple analogy in the way **forces** and **torques** cause the motion of objects. 

Just as, 

$$\sum_{i}\vec{f_{i}}=m\vec{a}$$

we have,

$$\sum_{i}\vec{\tau_{i}}=I\vec{\alpha}$$

where $I=mr^{2}$ (for a point mass) and $\vec{\alpha}=\frac{d\vec{\omega}}{dt}$.

$I$ is called the inertia. Just like for a given force, a large mass gives you a low **linear** acceleration (of the center of mass), for a given torque, a large inertia leads to a low **angular** acceleration (\vec{\alpha}). It quantifies how difficult it is to make an object rotate (change its angular coordinates $\theta$). In this course, the inertia (that is supposed to be calculated will always be **given** to you. 

Just as the force is the time derivative of momentum $\vec{p}$ (sometimes called the **linear** momentum), there exists a so-called **angular momentum**, noted $\vec{L}$. We have that the total time derivative of $\vec{L}$ equals the vector sum of all the torques applied at the same instant $t$ to the object.

$$\frac{d\vec{L}}{dt}=\sum_{i}\vec{\tau_{i}}=I\vec{\alpha}$$

The definition of angular momentum for a single point particle is:

$$\vec{L} =\vec{r} \times \vec{p}$$

where $\vec{p}$ is the particle's linear momentum and $\vec{r}$ is the position vector from the origin. The time-derivative of this is:

$$\frac{d\vec{L}}{dt}=\vec{r}\times\frac{d\vec{p}}{dt}+\frac{d\vec{r}}{dt}\times\vec{p}$$

This result can be easily be proven by splitting the vector into components and applying the **chain rule**. Novec{p}w,dt using the definition of the **net** force ($\vec{F}=\frac{d\vec{p}}{dt}$ (with this definition, the mass can vary) and the definition of instantaneous velocity $\vec{v}=\frac{d\vec{r}}{dt}$, we have:

$$\frac{d\vec{L}}{dt}=\vec{r}\times\vec{F}+\vec{v}\times\vec{p}$$.

The cross product of the **linear** momentum $\vec{p}$ and its associated velocity $\vec{v}$ is zero (because velocity and momentum are colinear - the area of the parallelogram is zero).

By definition, the term $\vec{r}\times\vec{F}$ is a torque. It occurs then that:

Torque on a particule is equal to the time derivative of its **angular** momentum. This is a general proof for point particules.

$$\vec{t}=-t\;\textbf{e}_{x'}$$

The sum of the forces in the tilted frame (the net force acting on the circle) is:

$$\vec{f}+\vec{n}+\vec{t}=(-mgsin(\phi)-t)\;\textbf{e}_{x'}+(n-mgcos(\phi))\;\textbf{e}_{y'}$$

And this net force is equal to the mass times the acceleration of the center of mass.

$$a_{x}^{CM}=\frac{1}{m}(-mgsin(\phi)-t)=R\frac{d^{2}\theta(t)}{dt^{2}}$$

and

$$a_{y}^{CM}=\frac{1}{m}(n-mgcos(\phi))=0$$

(These are **scalar** equations).

Now, what we have to note to go further is that there are **torques** at play in this motion. **Forces** are only part of the story. 

How do we compute the sum of the torque? One have to choose a reference point. Here: The center of mass is chosen.

From this point, the **sum** of the torques is:

$$\sum_{i}\vec{\tau_{i}}=\vec{0}\times m\vec{g}+(-R\;\textbf{e}_{y'})\times(-t\;\textbf{e}_{x'})=-Rt\textbf{e}_{z}$$

You see that the **torque** associated with the gravitationnal force is **zero** (for this specific choice of point of reference) and the net torque is a vector pointing inside your screen ($\textbf{e}_{y'}\times\textbf{e}_{x'}=-\textbf{e}_{z}$) whose magnitude is $Rt$. 

But, the sum of the torques is also equal to inertia times **angular acceleration**. We have:

$$-Rt\;\textbf{e}_{z}=-I\frac{d^{2}\theta(t)}{dt^{2}}\;\textbf{e}_{z}$$

Again, this extra minus sign comes from the fact that the circle turns clockwise. 

The inertia is **given** by:

- For the hoop:

$$I=mR^{2}$$

- For the disk:

$$I=\frac{1}{2}mR^{2}$$

This is of the form $bmR^{2}$ where $b$ is a constant with no physical dimensions (for the hoop, $(b= 1)$ and for the disk, ($b=\frac{1}{2}$). You can **compute** it for different geometries but in this course, it will always be **given** to you.

Ressembling all the equations together, we have:

$$-mgsin(\phi)-bma_{x}^{CM}(t)=ma_{x}^{CM}(t)$$

which gives:

$$a_{x}^{CM}(t)=\frac{gsin(\phi)}{(1+b)}$$

Hopefully, you find that it is a 1D **constant** acceleration motion. A time-dependent acceleration would have been completelly non-sensical because the source of the motion is the constant acceleration $g$. This answer tells you that: **The disk wins the race !** (and now, you know why). 

From this formula, you can integrate the acceleration and then velocity to get the time-dependent position.

<div class="alert alert-block alert-danger">
Caution: This is not a uniform circular motion because the angular velocity $\omega$ is not constant. It varies linearly with time. Do not make the confusion. The formulas for the uniform circular motion cannot be applied. 
</div>
